Começamos tendo que entender que o método da Bissecção considera a raiz dessa função, em que exista um x que resulte a função em 0, como delimitada em um intervalo dentro dos reais que vai de um número a até um número b. Depois vamos escrever a função:

O que confere na cálculadora. Vamos considerar nosso 'epson' = precisao

```python
import numpy as np
import math
import matplotlib.pyplot as plt

def f(x):
  return x**(3/2) - np.cos(x)

a, b = 0, 5000
precisao = 0.1 #nosso E
graf = []

if f(a)*f(b) < 0:
    while (b-a)/2 > precisao:
        media = (a+b)/2
        if f(a)*f(media) < 0:
            b = media
        else:
            a = media
        result = f(media)
        graf.append(result)
    print("A solução pelo método da Bissecção é: ",media)

plt.plot(graf)
plt.show()
```

```python
import math
import numpy as np

def f(x):
  return x**(3/2) - np.cos(x)

def df(x):
  return 3*x**(1/2)/2 + np.sin(x)

graf = []
erro = 100
x = 5000
max_calculos = 1000
i = 0
e = 0.0000001
graf.append(x)

while i <= max_calculos and erro > e:
  x_1 = x - f(x)/df(x)
  erro = abs(x_1-x)/abs(x_1)
  x = x_1
  i += 1
  graf.append(x)

if erro < e:
  print("A solução pelo método de Newton_Raphson é: ", x)
else:
  print("Número máximo de iterações excedido")

plt.plot(graf)
plt.show()
```
